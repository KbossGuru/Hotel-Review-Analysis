```python
#import the libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#load the dataset

hotel= pd.read_csv('Hotel_Reviews.csv')

hotel.head()
```

# Data Cleaning and Wrangling

```python
#delete the columns that are not needed

hotel = hotel.drop(['Total_Number_of_Reviews', 'Tags', 'days_since_review'], axis= 1)

#in the negative_reviews column, change the values for no negatives to 0 and the other entries to 1

hotel['Negative_Review']= np.where(hotel['Negative_Review'] == 'No Negative', 0, hotel['Negative_Review'])
hotel['Negative_Review']= np.where(hotel['Negative_Review'] != 0, 1, hotel['Negative_Review'])

#in the positive_reviews column, change the values for no positives to 0 and the other entries to 1
hotel['Positive_Review']= np.where(hotel['Positive_Review'] == 'No Positive', 0, hotel['Positive_Review'])
hotel['Positive_Review']= np.where(hotel['Positive_Review'] != 0, 1, hotel['Positive_Review'])

#change the names of some coluumns to more readable format
hotel.rename(columns = {'Review_Total_Negative_Word_Counts': 'Negative_Words',
             'Review_Total_Positive_Word_Counts' : 'Positive_Words', 
              'Total_Number_of_Reviews_Reviewer_Has_Given': 'Reviews_Given'}, inplace = True)

#change the datatype of the review date column to datetime
hotel['Review_Date'] = pd.to_datetime(hotel['Review_Date'])

#create a new column for Hotel_country
Country= []
for add in list(hotel['Hotel_Address']):
    Country.append(add.split()[-1])

hotel['Hotel_Country']= pd.DataFrame(Country)
hotel['Hotel_Country']= np.where(hotel['Hotel_Country']== 'Kingdom', 'United Kingdom', hotel['Hotel_Country'])
hotel['Hotel_Country']= np.where(hotel['Hotel_Country']== 'Netherlands', 'Netherland', hotel['Hotel_Country'])

hotel.head()
```

```python
#store the clean data in a csv file for visualization purposes

hotel.to_csv('hotel.csv')
```

# Exploratory Data Analysis

```python
#find the number of hotels per country
for country in hotel['Hotel_Country'].unique():
    count= hotel[hotel['Hotel_Country'] == country]
    hotels = len(count['Hotel_Name'].unique())
    print(country, hotels)
```

```python
#number of reviewers per country
for country in hotel['Hotel_Country'].unique():
    count= hotel[hotel['Hotel_Country'] == country]
    hotels = len(count['Hotel_Name'])
    print(country, hotels)
```

```python
#number of reviews per country
for country in hotel['Hotel_Country'].unique():
    count= hotel[hotel['Hotel_Country'] == country]
    hotels = sum(count['Total_Number_of_Reviews'].unique())
    print(country, hotels)
```

```python
#find the percentage of positive and negative reviews
positive = len(hotel[hotel['Positive_Review']== 1]['Positive_Review'])
negative = len(hotel[hotel['Negative_Review']== 0]['Negative_Review'])


print('Positive Reviews-'+ str(positive*100/(positive + negative)) )
print('Negative Reviews-' + str(negative*100/(positive + negative)) )
```

```python
#find the reviews per month

for month in hotel['Review_Date'].dt.month.unique():
    reviews = hotel[hotel['Review_Date'].dt.month==month]
    count = len(reviews['Review_Date'])
    print(f"{month}-{count}")
```

```python
#find the top 5 hotels with the most positive reviews
positive = hotel[hotel['Positive_Review']== 1]

numbers=[]
hotels=[]
for hotel in positive['Hotel_Name'].unique():
    review= positive[positive['Hotel_Name']==hotel]
    count= sum(review['Positive_Words'])
    hotels.append(hotel)
    numbers.append(count)
result=pd.DataFrame()
result['hotel'] = hotels
result['positive_words']=numbers
result.sort_values('positive_words', ascending=False).head(5)
```

```python
#find the top 5 hotels with the most negative reviews
negative = hotel[hotel['Negative_Review']== 1]

numbers=[]
hotels=[]
for hotel in negative['Hotel_Name'].unique():
    review= negative[negative['Hotel_Name']==hotel]
    count= sum(review['Negative_Words'])
    hotels.append(hotel)
    numbers.append(count)
result=pd.DataFrame()
result['hotel'] = hotels
result['negative_words']=numbers
result.sort_values('negative_words', ascending=False).head(5)
```

```python
#top 5 hotels with the highest average rating
hotels=[]
ratings=[]
for hosts in hotel['Hotel_Name'].unique():
    df= hotel[hotel['Hotel_Name']==hosts]
    rating= np.mean(df['Reviewer_Score'])
    hotels.append(hosts)
    ratings.append(rating)
result=pd.DataFrame()
result['hotel'] = hotels
result['Rating']=ratings
result.sort_values('Rating', ascending=False).head(5)
```

```python
#top 5 hotels with the least ratings
result.sort_values('Rating', ascending=True).head(5)
```

```python
#countries that has the highest positive ratings
countries=[]
ratings=[]
for country in hotel['Hotel_Country'].unique():
    df= hotel[hotel['Hotel_Country']==country]
    rating= np.mean(df['Reviewer_Score'])
    countries.append(country)
    ratings.append(rating)
result=pd.DataFrame()
result['Country'] = countries
result['Rating']=ratings
result.sort_values('Rating', ascending=False)
```

```python
#countries that produced the most reviewers
Country=[]
Reviewers=[]
for country in hotel['Reviewer_Nationality'].unique():
    count= hotel[hotel['Reviewer_Nationality'] == country]
    reviews = len(count['Hotel_Name'])
    Country.append(country)
    Reviewers.append(reviews)
result=pd.DataFrame()
result['Country'] = Country
result['Reviewers']=Reviewers
result.sort_values('Reviewers', ascending=False).head(5)
```
